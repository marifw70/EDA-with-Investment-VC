```python
#untuk menvisualisasikan data yang akan diolah, sehingga bisa tau dimana saja letak NAnnya
plt.figure(figsize=(20,20))# untuk size
sns.heatmap(data.isna()) # untuk plot heatmap NaN
plt.show()
# plt.savefig('handling missing.png') # untuk mensave hasil plot
```

### 7. Dropna & Fillna

```python
# Dari pengecekan column jenis kategori, terdapat beberapa Nan, 
#karena kita tidak tahu jenis categori untuk perusahaan yang tidak memiliki categori, maka datanya akan didrop
data.dropna(subset = ['category_list'], inplace= True)
data
```

```python
#pengecekan juga terhadap market
data.dropna(subset= ['market'], inplace = True)
data
```

```python
# data pada funding total formatnya masi ada koma dan salah sehingga perlu adanya replace pada values funding total
data['funding_total_usd'] = data['funding_total_usd'].str.strip().str.replace(",","")
data['funding_total_usd'] = data['funding_total_usd'].replace("-",0).astype("float")
data
```

###  D. Outliers :

```python
data.describe()['funding_total_usd']
```

```python
#untuk mendefine Q3 danQ1 dari funding total usd
Q3_funding = data.describe()['funding_total_usd']['75%']
Q1_funding = data.describe()['funding_total_usd']['25%']
```

```python
#untuk mencari IQR dari funding total usd
IQR = Q3_funding-Q1_funding

IQR
```

IQR = Q3-Q1

```python
#untuk mencari upper fence
UpperFence = Q3_funding + (1.5*IQR)

UpperFence
```

```python
#untuk mencari lower fence
LowerFence = Q1_funding - (1.5*IQR)

LowerFence
```

```python
#untuk menampilkan data outlier
Outliers = data[(data['funding_total_usd'] < LowerFence) | (data['funding_total_usd'] > UpperFence)]

Outliers.sort_values(by= 'funding_total_usd', ascending=False).head()
```

```python
#menampilkan boxplot dari funding_total_usd
plt.boxplot(Outliers['funding_total_usd'])
plt.show()
```
